Demo r√°pido y funcional del proyecto de an√°lisis de c√°ncer.

```python
# Importaciones b√°sicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

print("‚úÖ Importaciones exitosas")
print(f"üìÖ {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

```python
# Generar datos de ejemplo
np.random.seed(42)

data = {
    'Tipo': ['Pulm√≥n', 'Mama', 'Pr√≥stata', 'Colon'],
    'Casos': [150, 120, 90, 110],
    'Malignos': [85, 45, 35, 60]
}

df = pd.DataFrame(data)
df['Benignos'] = df['Casos'] - df['Malignos']
df['Tasa_Malignidad'] = df['Malignos'] / df['Casos']

print("üìä Datos generados:")
print(df)
```

```python
# Visualizaci√≥n
plt.figure(figsize=(12, 5))

# Gr√°fico 1
plt.subplot(1, 2, 1)
plt.bar(df['Tipo'], df['Casos'], color='skyblue', alpha=0.7)
plt.title('Total de Casos por Tipo')
plt.xlabel('Tipo de C√°ncer')
plt.ylabel('Casos')
plt.xticks(rotation=45)

# Gr√°fico 2
plt.subplot(1, 2, 2)
x = range(len(df))
plt.bar(x, df['Benignos'], label='Benigno', color='lightgreen', alpha=0.7)
plt.bar(x, df['Malignos'], bottom=df['Benignos'], label='Maligno', color='salmon', alpha=0.7)
plt.title('Distribuci√≥n Benigno vs Maligno')
plt.xlabel('Tipo de C√°ncer')
plt.ylabel('Casos')
plt.xticks(x, df['Tipo'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

print("‚úÖ Visualizaciones generadas")
```

```python
# Estad√≠sticas
print("üìä ESTAD√çSTICAS:")
print(f"Total casos: {df['Casos'].sum()}")
print(f"Total malignos: {df['Malignos'].sum()}")
print(f"Tasa promedio malignidad: {df['Tasa_Malignidad'].mean():.2%}")
print(f"Tipo m√°s agresivo: {df.loc[df['Tasa_Malignidad'].idxmax(), 'Tipo']}")

print("\nüéâ ¬°An√°lisis completado exitosamente!")
```
